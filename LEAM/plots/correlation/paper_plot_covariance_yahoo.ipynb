```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sklearn.preprocessing as sp
import seaborn as sns
```

```python
m = np.load('covariance.npy')
```

```python
class_name = ['Society Culture',
            'Science Mathematics',
            'Health' ,
            'Education Reference' ,
            'Computers Internet' ,
            'Sports' ,
            'Business Finance' ,
            'Entertainment Music' ,
            'Family Relationships' ,
            'Politics Government']
```

```python
plt.imshow(m, cmap='Blues',)
plt.colorbar()
plt.grid(False)
ax = plt.gca();
ax.xaxis.tick_top()
ax.set_xticks(np.arange(0, 10, 1));
ax.set_yticks(np.arange(0, 10, 1));
ax.set_xticklabels(class_name, rotation = 45, ha="left",);
ax.set_yticklabels(class_name);
plt.savefig('Covariance_matrix_yahoo_label.pdf', bbox_inches='tight')
```
