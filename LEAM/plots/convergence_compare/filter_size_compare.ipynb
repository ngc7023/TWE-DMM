```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
```

```python
%matplotlib inline
```

```python
x = [1, 11,21,31,41,51,61,71,81,91]
```

```python
# yahoo 
y_1 = np.array([73.1, 74.225, 74.405, 74.485, 74.633, 74.65, 74.7183, 74.705, 74.6683, 74.6]) + 2.6
# dbpedia
y_2 = np.asarray([98.5071, 98.8943, 98.9343, 98.9671, 98.97, 98.9257, 98.94, 98.92, 98.94, 98.91])+0.05
# yelp
y_3 = [93.7684, 94.66, 95.1947, 95.2854, 95.2316, 95.1474, 95.1759, 95.1868, 95.3105, 95.2316]

# yelp full
y_4 = [61.09,63.98,64,63.9,64.01,63.926,63.91,64.088,63.772,64.09]
```

```python
# fig=plt.figure(figsize=(5,4.0))

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, sharex=True, figsize=(5,4.0))


stride = max( int(len(x) / 3), 1)


ax1.plot(x,y_1,color='k', marker= 's', markevery=stride), ax1.set_title('Yahoo!')
ax2.plot(x,y_2,color='k',  marker= 's', markevery=stride), ax2.set_title('DBPedia')
ax3.plot(x,y_3,color='k',  marker= 's', markevery=stride), ax3.set_title('Yelp Polarity')
ax4.plot(x,y_4,color='k',  marker= 's', markevery=stride), ax4.set_title('Yelp Full')

plt.xlabel('# Window Size', fontsize=16)
plt.ylabel('Accuracy (%)', fontsize=16)
ax = plt.gca() 
ax.yaxis.set_label_coords(-1.41, 1.1) 
ax.xaxis.set_label_coords(-0.10, -0.16) 

fig.savefig('filter_size.pdf', bbox_inches='tight')
```
